# Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.
- Activation functions introduce non-linearity into neural networks.
Without them, a network would just be a linear combination of inputs and weights, meaning it couldn't learn complex patterns.

    They decide whether a neuron should “fire” or not.

    They allow networks to approximate non-linear relationships (like images, speech, or text).

    They help control the flow of information and gradients during training.
  
  - Sigmoid outputs values between 0 and 1 and is good for probabilities but suffers from vanishing gradients.
   - Tanh outputs between -1 and +1, is zero-centered, and better for hidden layers but still faces vanishing gradients.
   -  ReLU outputs 0 for negatives and the input itself for positives, making it fast and efficient, though it can cause “dying neurons.



# Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.
- A loss function measures the error for a single training example. It tells us how far the model's prediction is from the actual target for that one data point. For example, if a model predicts that a student will score 80 marks but the actual score is 90, the loss function calculates the error for that single prediction. In classification, if the model predicts “dog” for an image that is actually a “cat,” the loss function captures that mismatch for that one sample.

- A cost function, on the other hand, represents the overall performance of the model by combining the loss values across the entire dataset. It is usually the average of all individual losses. For instance, if we have exam scores for an entire class, the cost function would calculate the average error across all students' predictions. Similarly, in classification, the cost function evaluates how well the model performs across all images in the dataset, not just one.

# Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.  
- Optimizers adjust the weights and biases of a neural network to minimize errors and improve performance. They guide the training process, deciding how quickly and effectively the model learns.

- Gradient Descent: The simplest optimizer; updates weights step by step in the opposite direction of error. Easy to understand but can be slow and sensitive to learning rate.

- Adam: Combines momentum and adaptive learning rates. Fast, efficient, and widely used for large datasets. It often gives better results but can require more hyperparameter tuning.

- RMSprop: Improves training speed by adapting the learning rate for each parameter. Works well for recurrent networks and non-stationary problems, but still needs careful tuning.


# Question 6: Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate

```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

w = np.zeros(2)
b = 0
lr = 0.1

for _ in range(10):
    for i in range(len(X)):
        y_pred = 1 if np.dot(X[i], w) + b >= 0 else 0
        w += lr * (y[i] - y_pred) * X[i]
        b += lr * (y[i] - y_pred)

print("Weights:", w)
print("Bias:", b)
```

# Question 8: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

plt.plot(x, 1/(1+np.exp(-x)), label='Sigmoid')
plt.plot(x, np.maximum(0,x), label='ReLU')
plt.plot(x, np.tanh(x), label='Tanh')

plt.legend()
plt.show()
```

# Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=5, batch_size=32)

print("Training Accuracy:", history.history['accuracy'][-1])
```

# Question 9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.

```python
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

- Interpretation:

    Training loss decreases and accuracy increases, showing effective learning.

    Validation curves closely follow training curves, indicating good generalization.

    If validation loss started increasing, it would indicate overfitting, but here the model is stable and well-trained.

    Dropout layers
